```python
# What is the Question? What is the Hypothese? 
# What do we want to find out from the data? 
# This Example: IS AGE DEPENDENT ON LOCATION?
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# import data
user_data = pd.read_csv('/home/billyb/Downloads/git/datascience/user_data.csv')
pop_data = pd.read_csv('/home/billyb/Downloads/git/datascience/pop_data.csv')
```

```python
# merge tables
new_df = pd.merge(user_data, pop_data)

# determine rural or urban population
new_df.loc[new_df.population_proper < 100000, "location"] = "rural"
new_df.loc[new_df.population_proper >= 100000, "location"] = "urban"

# graph render adjustments
 
```

```python
# Paste histogram code:
age = new_df["age"]
sns.displot(age)
plt.show()

# Paste mean age location code:
location_mean_age = new_df.groupby("location").age.mean()
print(location_mean_age)

# Paste barplot code:
plt.close()
sns.barplot(
    data=new_df,
    x= "location",
    y= "age"
)
plt.show()

# Paste violinplot code:
plt.close()
sns.violinplot(x="location", y="age", data=new_df)
 
plt.show()
```

```python
# scatter plot of data points to understand the relationship of 
x = new_df["population_proper"]
y = new_df["age"]

plt.scatter(x, y, alpha=.15)
```

```python
sns.regplot(x="population_proper", y="age", data=new_df)
```
